```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib import cm

fig, ax = plt.subplots(1, figsize=(8, 8))

# the data
blood_lead_shp = 'child_blood_lead_levels_by_ct.shp'
blood_lead_gdf = gpd.read_file(blood_lead_shp)

# the color bar 
cmap = plt.cm.get_cmap('YlOrRd', 6)

# the map
blood_lead_gdf.plot(column='num_bll_5p', 
                         cmap='YlOrRd', 
                         edgecolor='0.5',
                         ax = ax,
                         linewidth=0.5,
                         scheme='equal_interval') 


norm = Normalize(vmin=blood_lead_gdf['num_bll_5p'].min(), vmax=blood_lead_gdf['num_bll_5p'].max())
n_cmap = cm.ScalarMappable(norm=norm, cmap=cmap)

# n_cmap.set_array([])
cbar = ax.get_figure().colorbar(n_cmap, orientation='vertical', fraction=0.02, pad=0.0, shrink=0.5)
cbar.ax.set_xlabel('number of children with blood lead level over 5 ug/dl')

ax.set_axis_off()
plt.axis('equal')
plt.show()
```
